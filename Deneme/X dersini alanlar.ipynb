```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

file = r'C:\Users\umut\Desktop\Proje\FirstPart_200801.xlsx'
df = pd.read_excel(file, sheet_name='Sheet1')

df=df.drop(columns=['Term','Max Capacity','Enrollment'])

df['Subj'] =df['Subj'].astype(str) + ' ' + df['Numb'].astype(str) #delete space
df.drop(columns=['Numb'],inplace=True)

sort_by_ID=df.sort_values('ID')
sort_by_ID=sort_by_ID.reset_index(drop=True)

df
```

```python
new_df = pd.DataFrame(columns=['ID','Student_Level','Subj','Section'])
rowCount=len(df.index)
for i in range (rowCount):
    if(df.iat[i,2]=='MATH 101'):
        id_stu=df.iat[i,0]
        for k in range (rowCount):
            if(i != k and df.iat[k,0]==id_stu):
                new_df=new_df.append(df.loc[[k]],ignore_index=True) 
                
new_df
```

```python
#her dersin kaçar kişi almış?
new_df=new_df.sort_values('Subj')
new_df=new_df.reset_index(drop=True)
new_df
```

```python
list_sub_nms=[]                      #dictinary mi yapsam?
list_ppl_count=[]
prev=''

for i in range (len(new_df)):
    x= new_df.iat[i,2]
    if(x != prev):
        list_sub_nms.append(x)
        list_ppl_count.append(1)
        prev=x
    else:
        temp=list_sub_nms.index(x)
        list_ppl_count[temp]=list_ppl_count[temp]+1
        
Combiined_List= pd.DataFrame ({'Subj': list_sub_nms,'Total_Students':list_ppl_count})
Combiined_List=Combiined_List.sort_values('Total_Students',ascending=False)
Combiined_List=Combiined_List.reset_index(drop=True)
Combiined_List
    
```
