```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

file = r'C:\Users\umut\Desktop\Proje\FirstPart_200801.xlsx'
df = pd.read_excel(file, sheet_name='Sheet1')

df=df.drop(columns=['Term','Max Capacity','Enrollment'])

df['Subj'] =df['Subj'].astype(str) + ' ' + df['Numb'].astype(str) #delete space
df.drop(columns=['Numb'],inplace=True)

sort_by_ID=df.sort_values('ID')
sort_by_ID=sort_by_ID.reset_index(drop=True)

df
```

```python
sort_by_Subj=df.sort_values('Subj')
subj_nm=[sort_by_Subj.iat[0,2]]
prev=subj_nm[0]
rowCount=len(df.index)

for i in range (rowCount):
    if(prev != sort_by_Subj.iat[i,2]):
        prev=sort_by_Subj.iat[i,2]
        subj_nm.append(prev)
subj_nm
```

```python
#2 lilerin hepsini çıkar

tempID=sort_by_ID.iat[0,0]
myArray = [[0] * len(subj_nm) for i in range(len(subj_nm))]

for i in range(rowCount):
    first_i=i
    while(i!=rowCount-1 and sort_by_ID.iat[i+1,0] == tempID):
        i=i+1
        temp1=subj_nm.index(sort_by_ID.iat[first_i,2])
        temp2=subj_nm.index(sort_by_ID.iat[i,2])
        myArray[temp1][temp2]= myArray[temp1][temp2] + 1
        myArray[temp2][temp1]= myArray[temp2][temp1] + 1
        #add 1 to first_i subj x i subj   +   i subj x first_i subj
        
    
    
    if(i!=rowCount-1  and  sort_by_ID.iat[first_i,0] != sort_by_ID.iat[first_i+1,0]):
        tempID=sort_by_ID.iat[first_i+1,0]
```

```python
pairs_index=[]
pairs_numOfStu=[]

def my_index(obj, L):
    try:
        return L.index(obj)
    except ValueError:
        return -1

    
tempID=sort_by_ID.iat[0,0]

for i in range(rowCount):
    first_i=i
    index_first=str(subj_nm.index(sort_by_ID.iat[first_i,2]))
    
    while(i!=rowCount-1 and sort_by_ID.iat[i+1,0] == tempID):
        i=i+1
        index_second=str(subj_nm.index(sort_by_ID.iat[i,2]))
        
        try1=my_index((index_first+','+index_second),pairs_index)
        try2=my_index((index_second+','+index_first),pairs_index)
        
        if( try1 ==-1 and try2==-1):
            pairs_index.append((index_first+','+index_second))   #string indexfirst +','+string indexsecond
            pairs_numOfStu.append(1)
        else:
            if(try1 != -1):
                pairs_numOfStu[try1]=pairs_numOfStu[try1]+1
            else:
                pairs_numOfStu[try2]=pairs_numOfStu[try2]+1
    
    
    if(i!=rowCount-1  and  sort_by_ID.iat[first_i,0] != sort_by_ID.iat[first_i+1,0]):
        tempID=sort_by_ID.iat[first_i+1,0]
    
list_= pd.DataFrame ({'Pairs':pairs_index,'Total_Students':pairs_numOfStu})
list_
```

```python
list_.sort_values('Total_Students',inplace=True,ascending=False)
list_.reset_index(drop=True,inplace=True)

list_
```
