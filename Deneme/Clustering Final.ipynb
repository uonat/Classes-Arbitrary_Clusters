```python
def X_Taken_students(df_ , Subj):
    new_df = pd.DataFrame(columns=['ID','Student_Level','Subj','Section'])
    rowCount=len(df_.index)
    for i in range (rowCount):
        if(df_.iat[i,2]== Subj):
            id_stu=df_.iat[i,0]
            for k in range (rowCount):
                if(i != k and df_.iat[k,0]==id_stu):
                    new_df=new_df.append(df_.loc[[k]],ignore_index=True) 
    new_df=new_df.sort_values('Subj')
    new_df=new_df.reset_index(drop=True)                
    return new_df

def Classes_Students(new_df_):           
    list_sub_nms=[]                      #dictinary mi yapsam?
    list_ppl_count=[]
    prev=''

    for i in range (len(new_df_)):
        x= new_df_.iat[i,2]
        if(x != prev):
            list_sub_nms.append(x)
            list_ppl_count.append(1)
            prev=x
        else:
            temp=list_sub_nms.index(x)
            list_ppl_count[temp]=list_ppl_count[temp]+1

    Combiined_List= pd.DataFrame ({'Subj': list_sub_nms,'Total_Students':list_ppl_count})
    Combiined_List=Combiined_List.sort_values('Total_Students',ascending=False)
    Combiined_List=Combiined_List.reset_index(drop=True)
    return Combiined_List
```
